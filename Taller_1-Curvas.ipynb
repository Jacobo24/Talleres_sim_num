1. Cree una función en Python que, a partir de un conjunto de datos $x_i$, determine el valor promedio de los datos, la desviación estándar, la varianza y el coeficiente de variación.

```python
import numpy as np

# Datos proporcionados
datos = np.array([
    6.495, 6.595, 6.615, 6.635, 6.485, 6.555, 6.665, 6.505,
    6.435, 6.625, 6.715, 6.655, 6.755, 6.625, 6.715, 6.575,
    6.655, 6.605, 6.565, 6.515, 6.555, 6.395, 6.775, 6.685
])

# Función para calcular estadísticas
def calcular_estadisticas(datos):
    media = np.mean(datos)
    desviacion_std = np.std(datos, ddof=1)  # ddof=1 para muestra
    varianza = np.var(datos, ddof=1)
    coef_variacion = (desviacion_std / media) * 100

    return {
        "Media": media,
        "Desviación estándar": desviacion_std,
        "Varianza": varianza,
        "Coeficiente de variación (%)": coef_variacion
    }

# Calcular estadísticas
estadisticas = calcular_estadisticas(datos)
estadisticas
```
